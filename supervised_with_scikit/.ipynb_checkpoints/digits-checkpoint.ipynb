```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets

import warnings
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

%matplotlib notebook
```

```python
digits = datasets.load_digits()
```

```python
X = digits.data
y = digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=49, stratify=y)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    ax.text(0, 7, str(digits.target[i]))
```

```python
training_score = np.empty(10)
test_score = np.empty(10)
passes = np.arange(1, 11)
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    training_score[i-1] = knn.score(X_train, y_train)
    test_score[i-1] = knn.score(X_test, y_test)
```

```python
plt.figure()
plt.plot(passes, training_score)
plt.plot(passes, test_score)
plt.show()
```
