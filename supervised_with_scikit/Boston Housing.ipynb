```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")
plt.style.use('seaborn')
%matplotlib notebook
```

```python
housing = datasets.load_boston()
```

```python
data = pd.DataFrame(housing.data, columns=housing.feature_names)
target = pd.DataFrame(housing.target, columns=['MEDV'])
```

```python
data.head()
```

```python
target.head()
```

```python
X = data.values
y = target.values
```

```python
X_rooms = X[:, 5]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_rooms, y, color='green')
plt.ylabel('Value of house / 1000$')
plt.xlabel('Number of rooms')
plt.grid(True, linestyle='-.')
plt.show()
```

```python
# Using linear regression model
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(X_rooms.reshape(-1,1), y)
pred_space = np.linspace(min(X_rooms), max(X_rooms)).reshape(-1, 1)
```

```python
plt.figure()
plt.scatter(X_rooms, y, color='red')
plt.plot(pred_space, reg.predict(pred_space), color='black', linewidth=2, linestyle='--')
plt.ylabel('Value of house / 1000$')
plt.xlabel('Number of rooms')
plt.grid(True, linestyle='-.')
plt.show()
```
